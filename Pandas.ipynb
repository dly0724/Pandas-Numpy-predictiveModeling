```python
# MAKING CHANGES TO DATA #

# Get total number of Speed, generation, legendary of the first 5 index from 0 to 4
df['Total'] = df['Speed'] + df['Generation'] + df['Legendary']
print(df['Total'].head(5))
# Hint: after the total is out, it is always better to double check if the total is the right number
# otherwise, it is very dangerous.
```

```python
# Drop a column
df.drop(columns=['Total'])
```

```python
# A better way to get total number of specific columns using iloc. sum method allows
# us to sum them up. axis is 1 means sum up horizontally. if axis is 0 means sum up
# vertiacally
# Notice the end parameter is inclusive for the first number, but exclusive for the last
# one. So in order to get the 9th column, we need to put a 10th.
df['Total'] = df.iloc[:, 4:10].sum(axis=1)
print(df['Total'])
```

```python
# Reorder column. Get a list of columns
columns = list(df.columns)
# Want the first four columns, then reverse indexing to get the total column at the end to
# after index 3 column, then the rest
df = df[columns[0:4] + [columns[-1]] + columns[4:12]]
print(df)
# Hint: be careful about the hard coded numbers, because it is very easy to mess things up
# if the data is changing. Safer to just type in the name of the column. Afterall,
# always check are the numbers correct.
```

```python
# Filter out a certain name. For example, name that has "Mega" in it
# use str.contains method, and use the "~" squiggly line to filter out
print(df.loc[~df['Name'].str.contains('Mega')])
# Hint: str.contains is very powerful in filtering
```

```python
# After importing regular expression (re) (A regular expression specifies a set of strings that matches it;
# the functions in this module let you check if a particular string matches a given regular expression)
print(df.loc[df['Type 1'].str.contains('Fire|Grass', regex=True)])
```

```python
# Hint: to ignore the capitalization problem of the words, use flags=re.I which ignores the
# letter's capitalization
print(df.loc[df['Type 1'].str.contains('fire|grass', regex=True, flags=re.I)])
```

```python
# Find the 0 or more name that contains the string 'pi' that follows with anything from a to z after it
# ignore the captilization again by using regular expression re.I
print(df.loc[df['Name'].str.contains('pi[a-z]*', regex=True, flags=re.I)])
```

```python
# ^ means start of line, if wanna get the name string that starts with pi.
print(df.loc[df['Name'].str.contains('^pi[a-z]*', regex=True, flags=re.I)])
```

```python

# AGGREGATE STATISTICS (Groupby) #
# used to split the data into groups based on some criteria #

# Get all the data broken down by their mean sorted by what type 1 is
df = pd.read_csv('modified.csv')
print(df.groupby(['Type 1']).mean())
# Hint: three things that you can use in groupby are: mean, sum and count.
```

```python
# Get all the data broken down by their mean sorted by what type 1 is. Sort
# the value of speed from highest to the lowest
print(df.groupby(['Type 1']).mean().sort_values('Speed', ascending=False))
```

```python
# adding a count column, assign each row's count as 1
#df['count'] = 1
# Only get the group by type 1 count column
print(df.groupby(['Type 1']).count()['count'])
```

```python
# Of the type 1 group to have the type 2 group
# it is useful if working with really massive data set, when you could not even
# load them all into one data frame
print(df.groupby(['Type 1', 'Type 2']).count()['count'])
```

```python
# WORKING WITH LARGE AMOUNTS OF DATA #
# Imagining working with a file that is 20 gigabytes big #
# Pandas help to read the a chunk of the file at a time #
# Then you do not have to load them all to the memory #

# Chunksize allows 5 rows to be passed in at a time
for df in pd.read_csv('modified.csv', chunksize=5):
    print("CHUNK DF")
    print(df)
```

```python
# Define new empty data frame with same column name
new_df = pd.DataFrame(columns = df.columns)
for df in pd.read_csv('modified.csv', chunksize=5):
    results = df.groupby(['Type 1']).count()

    # Append two data frame together
    # As we do more iterations, we are building this new data frame of all info in orignial dataset
    # because each chunk we are doing group by and count, the data size is being shrinked down
    # so the new data frame is a lot smaller, can do more analysis on this smaller one
    new_df = pd.concat([new_df, results])
```
